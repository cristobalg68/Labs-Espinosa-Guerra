<code>Escriba sus respuestas acá:</code>

1. Git es un Sistema de Control de Versiones Distribuido (DVCS por sus siglas en inglés). Este, a diferencia de los otros tipos de Sistemas de Control de Versiones (los Locales y los Centralizados), permite que los clientes puedan replicar todo el repositorio de un proyecto, no solo la última versión de este. Además, cada clon del repositorio es una copia completa de todos los datos. También, a diferencia de los Sistemas de Control de Versiones Centralizados (CVCS por sus siglas en ingles), este tipo de VCS permite tener varios flujos de trabajo.

2. * Git es una herramienta de control de versiones distribuida, mientras que GitHub es una plataforma basada en la nube que incorpora las características de control de versiones de Git.

  * Git es una herramienta de código abierto que los desarrolladores instalan localmente para gestionar el código de forma local, mientras que GitHub es un servicio en línea al que los desarolladores que usan Git pueden cargar o descargar sus repositorios.

  * Git al estar instalado de forma local no necesita internet para su uso, mientras que GitHub, al ser un servicio en la nube, necesita de una conexión a internet (si bien existe una versión de escritorio de GitHub, esta no es necesaria). Además, Git puede ser utilizado sin GitHub, no en viceversa, pues GitHub necesita de Git.

3. Un Hash es un valor que se suele usar para identificar de forma única un elemento. En particular, el modelo de datos de Git utiliza el Algoritmo de Hash Seguro 1 (SHA-1 por sus siglas en inglés) que genera un Hash de 40 caracteres. Dicho valor Hash de 40 caracteres se utiliza para referenciar de forma única a un blop, un commit o un árbol. Si bien, el valor Hash es de 40 caracteres, se suelen utilizar solo unos pocos (6 caracteres) para identificar un commit. En la siguiente imagen se puede ver un ejemplo de lo anterior.

1.PNG

4. Los elementos que componen la estructura de datos de Git son el Blop, el Tree y el Commit. En un Blop se almacena todo el contenido de un archivo. Un Tree representa los directorios para un blop y para más trees. Y un Commit es la confirmación de algún cambio que se le haya realizado al repositorio.

5. Para crear un repositorio desde cero en nuestro entorno local se utiliza el comando `git init nombre-del-repositorio`. Por otro lado, si se quiere hacer seguimiento de los cambios, el comando de mayor utilidad es `git diff`, con el cual se pueden ver las diferencias entre conjuntos de archivos en el repositorio.



Responda si las siguientes afirmaciones son Verdaderas (V) o Falsas (F). Justifique en **cualquiera** de los casos (0.2 pts c/u):


6. `Git` siempre resuelve automáticamente los problemas de lineas al hacer merge entre diferentes ramas.

7. Ejecutar `git pull` sobre un repositorio es equivalente a ejecutar `git clone`.

8. Cada miembro del equipo trabaja en la misma copia del código fuente usando `Git`.

9. En `Git` es posible deshacer cambios y volver a versiones anteriores del código.

10. El comando `git commit` se utiliza para enviar cambios al repositorio remoto.

<code>Escriba sus respuestas acá:</code>

6. Falso, no siempre `Git` resuelve de manera automática estos problemas ya que cuando se juntan ramas con cambios conflictivos entre si, `Git` se detiene y requiere que el usuario resuelva este conflicto manualmente, pues `Git` no necesariamente puede saber qué cambios mantener y cuáles otros no hacerlo.

7. Falso, no son equivalentes ambos comandos: `git pull` se usa para actualizar la copia local ya existente solo con los cambios más recientes del repositorio remoto. Mientras que `git clone` se usa para obtener una copia completa del repositorio remoto, guardada en un entorno local, esta clonación incluye el historial completo, ramas, etiquetas y commits. 

8. Falso, cada miembro del equipo tiene su propia copia local del repositorio, sobre la cual puede trabajar. Cuando alguien del equipo realiza un cambio importante sobre su copia, puede crear un `pull request` para que el resto revise y comente el cambio antes de incorporarlo al repositorio principal.

9. Verdadero, `Git` almacena el historial completo de versiones de código fuente, esto permite deshacer cambios (mediante `git revert`) y volver a versiones antiguas (mediante `git checkout`) si así se desea. 

10. Falso, `git commit` se usa para guardar cambios en el repositorio local, no en el remoto. Para enviar cambios al repositorio remoto se utiliza `git push`.
